```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

```python
# Read Insurance CSV
df = pd.read_csv('insurance.csv')
print(df.head())
```

```python
# Draw a scatter plot to find relationship between age and insurance costs

plt.figure()
x = df.iloc[:,:1].values
y = df.iloc[:,-1].values
plt.scatter(x, y, color = 'b')
plt.xlabel('Age')
plt.ylabel('Insurance Charges')
plt.show()
```

```python
# Create model by running LinearRegression and fit the model
simple_model = LinearRegression()
simple_model.fit(x,y)
```

```python
# Plot the data and the best fit line based on the linear regression
y_pred = simple_model.predict(x)
plt.scatter(x, y, color = 'b')
plt.plot(x, y_pred, color = 'r')
plt.xlabel('Age')
plt.ylabel('Insurance Charges')
plt.show()
```
